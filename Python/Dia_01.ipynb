### Separadores:
- \n (quebra de linha)
- \t (tab)
- "," (virgula)
- ";" (ponto e virgula)

```python
sequencia = open("seq.fasta")
sequencia2 = sequencia.readlines()
```

```python
DNA = ""
for n in sequencia2:
    if n.startswith(">"):
        pass
    else:
        DNA = DNA + n   

DNA = DNA.replace("\n", "") 
```

```python
DNA
```

### Counting DNA Nucleotides

```python
print("Número de As: ", DNA.count("A"))
print("Número de Cs: ", DNA.count("C"))
print("Número de Gs: ", DNA.count("G"))
print("Número de Ts: ", DNA.count("T"))

percentGC = (DNA.count("G") + DNA.count("C")) / (DNA.count("G") + DNA.count("C") + DNA.count("A") + DNA.count("T"))

print("Percentual de GC: ", percentGC)
```

### Transcribing DNA into RNA

```python
RNA = DNA.replace("T", "U")
```
